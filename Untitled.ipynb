```python
# Convolution
toySignal = np.array([0, 1, 2, 1, 0])
toyKernel = np.array([1, 0, 1])

print "Signal: ", toySignal
print "Kernel: ", toyKernel

toyConvolution = []
for i in xrange(0, len(toySignal)-len(toyKernel) + 1):
    value = 0
    for j in xrange(len(toyKernel)):
        value += toySignal[i + j] * toyKernel[j]
        print toySignal[i + j], "*", toyKernel[j]
    toyConvolution.append(value)
    print

print "Convolution: ", toyConvolution
```

```python
from sklearn.datasets import load_sample_images
import scipy.signal
                      
# Keep just one of the channels to keep it simple
dataset = load_sample_images()
img = dataset.images[1] [:, :, 2] 

fig, ax = plt.subplots(1, 3)
fig.set_size_inches(15, 5)
ax[0].imshow(img, cmap='Greys')

# Goal find the vertical edges
ax[1].imshow(scipy.signal.convolve2d(img, np.array([[-1, 0, 1], 
              [-2, 0, 2], 
              [-1, 0, 1]]), 'valid'), cmap='Greys')

# Goal find the horizontal edges
ax[2].imshow(scipy.signal.convolve2d(img, np.array([[1, 2, 1], 
              [0, 0, 0], 
              [-1, -2, -1]]), 'valid'), cmap='Greys')

fig.tight_layout()
plt.show()
```
